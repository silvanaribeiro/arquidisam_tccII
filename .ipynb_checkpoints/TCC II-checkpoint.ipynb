# Coletando os Dados

```python
# encoding: utf-8
from __future__ import unicode_literals
import requests
from lxml import html

feam_url = "http://www.feam.br/noticias/1/1327-boletim-qualidade-do-ar"
page_content = requests.get(feam_url).content
html_tree = html.fromstring(page_content)
html_tree.resolve_base_href()
elements = html_tree.xpath('//a')
```

```python
possible_matches = []
for element in elements:
    link = element.get('href')
    if link is not None and ".xlsx" in link:
        print link
        possible_matches.append(link)

        
        
```

```python
from datetime import datetime
months = ['', 'Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
date = datetime.now()
```

```python
months[date.month]
```

```python
downloads = 0
for possible_match in possible_matches:
    if months[date.month] in possible_match:
        file_name = 'Dados/Dados-' + str(date.day) + '-' + str(date.month) + '-' + str(date.year) + '-' + str(downloads) + '.xlsx'
        result = requests.get(possible_match)
        downloads = downloads + 1
        with open(file_name, 'wb') as f:
            print(file_name)
            f.write(result.content)
        
```

# Ingestão dos dados de Poluição atmosférica

```python
class Polutant(object):
    def __init__(self, name, quantity, date=None):
        self.name = name
        self.quantity = quantity
        self.date = date
     
```

```python
import csv
from datetime import datetime
report_by_day = {}
first_row = True
date = None
with open('Dados/Poluentes.csv', 'rb') as f:
    reader = csv.reader(f)
    for row in reader:
        if first_row: # Skipping header row
            first_row = False
        else:
            first_col = True
            list_polutants = []
            qtd_polutants = 0
            for col in row:
                if first_col:
                    date = datetime.strptime(col, '%d-%m-%y') # first value is always the date
                    first_col = False
                elif col:
                    if not col.isdigit(): # if it's only numbers it's a polutants quantity
                        list_polutants.append(Polutant(col.strip(), 0))
                    else:
                        p = list_polutants[qtd_polutants]
                        p.quantity = int(col)
                        p.date = date
                        list_polutants[qtd_polutants] = p
                        qtd_polutants = qtd_polutants + 1
                        
           # report_by_day[date] = dict([(p.name, p.quantity) for p in list_polutants])
            report_by_day[date] = list_polutants
        
                    
                
```

```python
for date in sorted(report_by_day):
    print date.strftime('%d-%m-%y')
    for p in report_by_day.get(date):
        print p.name
        print p.quantity
        print p.date.strftime('%d-%m-%y')
    print '--------------'
```

# Exploração de dados de poluição

```python
max_polutants = []
for date in sorted(report_by_day):
    for p in report_by_day.get(date):
        if not any(x.name == p.name for x in max_polutants):
            max_polutants.append(p)
        else:
            for x in max_polutants:
                if x.name == p.name:
                    if x.quantity < p.quantity:
                        max_polutants.remove(x)
                        max_polutants.append(p)
                    else:
                        break
for p in max_polutants:
    print p.name
    print p.quantity
    print p.date.strftime('%d-%m-%y')
    print '--------------'        
    
```

# Preparação dos dados para aplicação do algoritmo K-means

```python
import pandas as pd
df_diseases = pd.DataFrame(diseases)
```

```python
df_diseases
```

```python
index = list(sorted(report_by_day.keys()))

columns = ['NO2', 'CO', 'O3', 'PTS', 'PI']
df_polutants = pd.DataFrame(index=index, columns=columns)
df_polutants = df_polutants.fillna(0) # with 0s rather than NaNs
```

```python
df_polutants.head(10)
```

```python
count = 0
for date in list(sorted(report_by_day.keys())):
    for p in report_by_day.get(date):
        if p.name == 'Dioxido de Nitrogenio':
            df_polutants.loc[date,'NO2'] = p.quantity
        elif p.name == 'Monoxido de Carbono':
            df_polutants.loc[date,'CO'] = p.quantity
        elif p.name == 'Ozonio':
            df_polutants.loc[date,'O3'] = p.quantity
        elif p.name == 'Particulas Totais em Suspensao':
            df_polutants.loc[date,'PTS'] = p.quantity
        elif p.name == 'Particulas Inalaveis':
            df_polutants.loc[date,'PI'] = p.quantity
```

```python
df_polutants.head(200)
```

```python
columns = ['Semana', 'Ano','NO2', 'CO', 'O3', 'PTS', 'PI']
df_polutants_by_week = pd.DataFrame(columns=columns)
df_polutants_by_week_max = pd.DataFrame(columns=columns)
df_polutants_by_week_mean = pd.DataFrame(columns=columns)
df_polutants_by_week_sum = pd.DataFrame(columns=columns)
```

```python
df_polutants_by_week
```

```python
curr_week = 0

for index, row in df_polutants.iterrows():
    df = pd.DataFrame([[str(index.isocalendar()[1]), str(index.year), int(row['NO2']), int(row['CO']), int(row['O3']), int(row['PTS']), int(row['PI'])]], columns=columns)
    df_polutants_by_week = df_polutants_by_week.append(df, ignore_index=True)
```

# Calculation of SSE (Sum of Square Error)

```python
import unicodecsv
import numpy


def calc_SSE(clusters):
    SSE = 0
    for cluster in clusters:
        meanCluster = [numpy.mean(numpy.array(list(cluster['A-caso'])).astype(numpy.float)),
                       numpy.mean(numpy.array(list(cluster['A-obito'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['B-caso'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['B-obito'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['H1N1-caso'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['H1N1-obito'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['H3-caso'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['H3-obito'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['SRAG-caso'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['SRAG-obito'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['NO2'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['CO'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['O3'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['PTS'])).astype(numpy.float)), 
                       numpy.mean(numpy.array(list(cluster['PI'])).astype(numpy.float))]
        for index, row in cluster.iterrows():
            point = numpy.array([row['A-caso'], row['A-obito'], row['B-caso'], row['B-obito'], row['H1N1-caso'], row['H1N1-obito'], row['H3-caso'], row['H3-obito'], row['SRAG-caso'], row['SRAG-obito'], row['NO2'], row['CO'], row['O3'], row['PTS'], row['PI']]).astype(numpy.float)
            SSE = SSE + sum((point-meanCluster)**2)
        
    print SSE
```
